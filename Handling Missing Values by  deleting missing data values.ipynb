```python
# FINDING PERCENTAGE OF MISSING VALUES
missing_values_percentage = df.isnull().sum()/df.shape[0]*100
missing_values_percentage
```

# Deleting Missing Data

Columns having missing rows more than 20 percentage of total rows should be deleted

```python
#drop_columns = missing_values_percentage[missing_values_percentage>20].keys()
# But since 17 % of column LotFrontage is near 20 % , take 17 % as cut off
drop_columns = missing_values_percentage[missing_values_percentage>17].keys()
drop_columns
```

```python
df_drop_col = df.drop(columns = drop_columns)
df_drop_col.shape
```

```python
sns.heatmap(df_drop_col.isnull())
```

Above heatmap shows that some rows in dataset are having missing values

```python
df_drop_row = df_drop_col.dropna()
sns.heatmap(df_drop_row.isnull())
```

Above heatmap shows that there is no missing values in the modified dataset : df_drop_row

```python
# Now select columns having numerical values to check whether this cleaning has adversly affected the data pattern
df_drop_row.select_dtypes(include=['int64','float64']).columns
```

```python
# Now check distance plot of this variable to check changes after dat cleaning

# Note Id won't put much affect on dataset...... so no need to take Id variable with other variabled for this evaluation

num_var = ['Id', 'MSSubClass', 'LotArea', 'OverallQual', 'OverallCond',
       'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2',
       'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF',
       'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
       'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces',
       'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF',
       'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal',
       'MoSold', 'YrSold', 'SalePrice']
```

```python
plt.figure(figsize=(25,25))
# use enumerate keyword to find index of item in the list
for j, var in enumerate(num_var):
    plt.subplot(9,4,j+1)
    sns.distplot(df[var], bins =20)
    sns.distplot(df_drop_row[var], bins =20)
```

Above plot shows that there is no much variation in numerical columns

```python
# Now focus categorical variables
df_drop_row.select_dtypes(include=['object']).columns
```

```python
# Creating function that shows % loss of data in various categoroes of categorical variables, same as done in case of numerical variables

def cat_var_dist(var):
    return pd.concat([df[var].value_counts()/df.shape[0] * 100, 
          df_drop_row[var].value_counts()/df_drop_row.shape[0] * 100], axis=1,
         keys=[var+'_org', var+'clean'])
    
```
